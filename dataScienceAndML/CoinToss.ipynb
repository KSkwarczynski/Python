```python
import random
import matplotlib.pyplot as plt
```

```python
#Basically 0-Tail and 1-Head
coin1 = random.randint(0,1)
coin2 = random.randint(0,1)
print("coin1 ",coin1," coin2 ",coin2)
```

```python
#Our F(E) is just a sum of coin result
FE = coin1 + coin2
```

```python
print(FE)
```

```python
#now CDF, this is bacically o set of if statement
E = random.randint(0,100)#could be floats but doesn't matter for this example
print(E)
```

```python
def CDF(x):
    if(x<1): return 1/4
    if(x>=1 and x<2): return 3/4 
    if(x>=2): return 4/4
```

```python
print(CDF(E))
#it's working as it should
```
