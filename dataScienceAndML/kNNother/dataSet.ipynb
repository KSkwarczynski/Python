```python
#some imports
import pandas as pd
#for visualization
import matplotlib.pyplot as plt
#to download file
import urllib.request
import math

%matplotlib inline
```

```python
url='http://koza.if.uj.edu.pl/~krzemien/machine_learning2021/materials/datasets/data1.csv'
urllib.request.urlretrieve(url,'data1.csv')
```

```python
input_file = 'data1.csv'
```

```python
data = pd.read_csv(input_file)
```

```python
#zapisuje nasze dane w postaci odzielnej listy dla x, y i dataSet
df = pd.DataFrame(data)
x = df['x'].tolist()
y = df['y'].tolist()
dataSet = df['dataSet'].tolist()
print(x)
print(y)
print(dataSet)
```

```python
#teraz chcemy aby kazdy data set mial oddzielna liste x i y, napewno da sie to ladniej zrobic, no ale trudno
X0 = []
X1 = []
X2 = []
X3 = []
Y0 = []
Y1 = []
Y2 = []
Y3 = []
for i in range(0, len(x)):
    if(dataSet[i] == 0):
        X0.append(x[i])
        Y0.append(y[i])
    if(dataSet[i] == 1):
        X1.append(x[i])
        Y1.append(y[i])
    if(dataSet[i] == 2):
        X2.append(x[i])
        Y2.append(y[i])
    if(dataSet[i] == 3):
        X3.append(x[i])
        Y3.append(y[i])
print(X0)
```

```python
#Teraz mamy dane uporzdkowane w latwy sposob to zaladuje funkcje ktore przygotwoalem w poprzednich zadaniach
def Mean(X):
    mean = 0
    for ic in range(0, len(X)):
        mean = mean + X[ic]
    mean = mean/len(X)
    return mean
def Variance(X):
    mean = Mean(X)
    variance = 0
    for ic in range(0, len(X)):
        variance = variance + math.pow(X[ic] - mean, 2)
    variance = variance/len(X)
    variance = math.sqrt(variance)
    return variance
def Cov(x,y):
    meanX = Mean(x)
    meanY = Mean(y)
    cov = 0
    for ic in range(0, len(x)):
        cov = cov + x[ic]*y[ic]
    cov = cov/len(x) - meanX*meanY 
    return cov
def Pearson(x,y):
    cov = Cov(x,y)
    varX = Variance(x)
    varY = Variance(y)
    Pearson = 0
    Pearson = cov/(varX*varY)
    return Pearson
```

```python
#Teraz obliczymy srednie, waraince dla wszystich "data sample"
MeanX = []
MeanY = []
VarX = []
VarY = []
COV = []
PEARSON = []

#To jest paskudny sposob aby to zrobic
MeanX.append(Mean(X0))
MeanX.append(Mean(X1))
MeanX.append(Mean(X2))
MeanX.append(Mean(X3))

MeanY.append(Mean(Y0))
MeanY.append(Mean(Y1))
MeanY.append(Mean(Y2))
MeanY.append(Mean(Y3))

#Wynik jest dziwny no ale trudno
print(MeanX)
print(MeanY)
```

```python
VarX.append(Variance(X0))
VarX.append(Variance(X1))
VarX.append(Variance(X2))
VarX.append(Variance(X3))

VarY.append(Variance(Y0))
VarY.append(Variance(Y1))
VarY.append(Variance(Y2))
VarY.append(Variance(Y3))

print(VarX)
print(VarY)
```

```python
COV.append(Cov(X0,Y0))
COV.append(Cov(X1,Y1))
COV.append(Cov(X2,Y2))
COV.append(Cov(X3,Y3))
print(COV)
```

```python
PEARSON.append(Pearson(X0,Y0))
PEARSON.append(Pearson(X1,Y1))
PEARSON.append(Pearson(X2,Y2))
PEARSON.append(Pearson(X3,Y3))
print(PEARSON)
```

```python
#Troche dziwne te zaleznosci no ale co zrobic
plt.plot( [0, 1, 2, 3], MeanX,
         [0, 1, 2, 3], MeanY)
```
