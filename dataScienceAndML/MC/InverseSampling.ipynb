```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#CDF dla rozkladu eksponencjalnego to CDF = 1 - exp(-x)
#Odwotnosc CDF to x = –ln(1–u)

N=10000
x = []
for i in range(N):
    u = random.uniform(0,1)
    x.append(-math.log(1-u));
    
plt.hist(x, 100)    
```
